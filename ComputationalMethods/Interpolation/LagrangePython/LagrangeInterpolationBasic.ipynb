```python
import LagrangeInterpolation as Lagrange
import matplotlib.pyplot as plt
```

```python
x_net = [1, 2, 3]
y_net = [1, 4, 9]
M = 100
K = 4
t = list(K*i/M for i in range(-M, M+1))
LP = Lagrange.LagrangeInterpolation(x_net, y_net)
```

```python
for k in range(len(x_net)):
    plt.plot(t, LP.basis_polynomial_arr(t, k), linestyle="--")

plt.plot(t, LP.interpolate_arr(t))
plt.show()
```
